```python
# %%

import matplotlib.pyplot as plt
from sklearn.metrics import ConfusionMatrixDisplay
import numpy as np
import seaborn as sns
# %%
plt.figure(figsize=(10, 7))
font = {'family' : 'normal',
        #'weight' : 'bold',
        'size'   : 15}

plt.rc('font', **font)

cf_matrix= [[45,  15],
       [ 15, 45]]

labels = ['True Negatives (Q1)','False Positives (Q2)','False Negatives (Q3)','True Positives (Q4)']
labels = np.asarray(labels).reshape(2,2)
sns.heatmap(cf_matrix, annot=labels, fmt='', cmap='Blues',
            xticklabels=['Model No','Model Yes'], yticklabels=['Data No','Data Yes'], cbar=False)

plt.savefig('..\..\Resultados\Confusion Matrix.png', dpi=600)
```
