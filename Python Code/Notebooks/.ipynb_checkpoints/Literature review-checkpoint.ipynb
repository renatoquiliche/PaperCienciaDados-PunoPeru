(Salman and Li, 2018) **Issues identified for future research include un-certainty propagation, decoupling changes in flood patterns caused by climate change and those caused by other factors such as populationand economic growth, consideration of infrastructure decay in future flood risk modeling, and consideration of the intangible benefits of risk mitigation.** **Climate will affect the future of disasters, disaster preparedness is important too**

The projection of the future trend of flood risk due to the impact ofclimate change and population and economic growth is marred byuncertainties. The sources of these uncertainties include: (i) futurehuman behavior, including future GHG emissions, policies to curbGHG emissions, technological advancements (in geoengineering,for example), and adaptation strategies; (ii) climate modeling, in-cluding uncertainty inherent in GCMs, uncertainty in modeling theimpact of human and natural climate driving factors on future cli-mate, and uncertainty in the downscaling of GCMs; (iii) hydrolog-ical models, including uncertainty due to insufficient or lack oflong-term data for hydrological modeling; (iv) hydraulic models;(v) flood damage models; and (vi) population and economic growthmodels. Among these sources, uncertainty in climate change mod-eling is relatively higher. A robust framework for modeling the fu-ture trend of flood risk needs to include uncertainty propagationamong the various risk assessment steps. Improved understandingof uncertainty is key to risk communication and policy decisionsthat will assign significant resources to adaptation strategies (Wilbyet al. 2008). 

A major obstacle to flood risk management is the lack of resour-ces (technical and financial) at the local governmental levels.Although some countries develop flood management strategiesat the national level, the often localized impact of floods makesit a local or state government affair. Even in countries that approachflood risk at the national level, implementation of policies devel-oped at the national level falls on local government authorities.As such, risk communication methods need to be refined to ensureproper communication with local government policy makers andinstitutions </div>

Proper public risk perception is key to effective flood riskmanagement. Poor public risk perception can jeopardize effectivemanagement strategies such as land-use policies and regulationsas well as insurance schemes. Furthermore, additional researchis needed to invent ways of overcoming the lack of interest frompublic and local planners in flood risk management. This includeseducation programs to point out the risks of potential floods(Godschalk et al. 2003). The use of visual aids such as photographsand videos to elucidate public flood risk perception can be exploredas part of the education program. Keller et al. (2006) argued thatusing visual aids such as negative photographs showing damagedue to natural hazards can be used to heighten the level of publicrisk perception. Another way to increase public participation is em-bedding flood risk management into comprehensive communityplanning. This will link risk mitigation with immediate quality-of-life concerns related to issues such as transportation, zoning,and land development.
